```python
import re

def regularization(text):
    text = text.lower()
    text = text.replace("@", '')
    text = " ".join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", text).split())
    return text
```

```python
regularization("@kotez #pulangkampung ke sedayu")
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
```

```python
def sastrawi_stem(text):
    stopword_factory = StopWordRemoverFactory()
    stopword_remover = stopword_factory.create_stop_word_remover()
    factory = StemmerFactory()
    stemmer = factory.create_stemmer()
    text = stopword_remover.remove(stemmer.stem(text))
    return text
```

```python
import nltk
nltk.download("stopwords")
from nltk.corpus import stopwords
```

```python
def nltk_stem(text):
    filtering = stopwords.words("indonesian", "english")
    words = text.split()
    filtered_words = []

    for word in words:
        if(word not in filtering):
            filtered_words.append(word)

    filtered_text = " ".join(filtered_words)
    return filtered_text
```

```python
df_anies['clean_text'] = df_anies['full_text'].apply(regularization).apply(sastrawi_stem).apply(nltk_stem)
df_anies['clean_text']
```

```python
df_ganjar['clean_text'] = df_ganjar['full_text'].apply(regularization).apply(sastrawi_stem).apply(nltk_stem)
df_ganjar['clean_text']
```

```python
df_prabowo['clean_text'] = df_prabowo['full_text'].apply(regularization).apply(sastrawi_stem).apply(nltk_stem)
df_prabowo['clean_text']
```

```python
anies_words = ''
for val in df_anies['clean_text']:

    # typecaste each val to string
    val = str(val)

    # split the value
    tokens = val.split()

    # Converts each token into lowercase
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()

    anies_words += " ".join(tokens)+" "
```

```python
from transformers import pipeline, AutoTokenizer, AutoModelForSequenceClassification

pretrained= "ayameRushia/roberta-base-indonesian-1.5G-sentiment-analysis-smsa"
model = AutoModelForSequenceClassification.from_pretrained(pretrained)
tokenizer = AutoTokenizer.from_pretrained(pretrained)

sentiment_analysis = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)
```

```python
sentiment_label = []
sentiment_score = []

for i in range (len(df_anies)):
  sent_indo = sentiment_analysis(df_anies.iloc[i]['clean_text'])
  label = sent_indo[0]['label'].lower()
  score = sent_indo[0]['score']
  sentiment_label.append(label)
  sentiment_score.append(score)

print(sentiment_label)
print(sentiment_score)

df_anies['label'] = sentiment_label
df_anies['score'] = sentiment_score
```

```python
# import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics

# Contoh DataFrame
# data = {'text': ['Ini adalah contoh positif', 'Ini contoh negatif', 'Ini contoh positif']}
# df = pd.DataFrame(data)
# df['label'] = ['positif', 'negatif', 'positif']

# Bagi data menjadi set pelatihan dan set pengujian
X_train, X_test, y_train, y_test = train_test_split(df['clean_text'], df['label_bert'], test_size=0.2, random_state=42)

# Vektorisasi data
vectorizer = CountVectorizer()
X_train_vect = vectorizer.fit_transform(X_train)
X_test_vect = vectorizer.transform(X_test)

# Inisialisasi dan pelatihan model Naive Bayes
nb_classifier = MultinomialNB()
nb_classifier.fit(X_train_vect, y_train)

# Prediksi
y_pred = nb_classifier.predict(X_test_vect)

# Evaluasi model (opsional)
accuracy = metrics.accuracy_score(y_test, y_pred)
print("Akurasi:", accuracy)
```

```python
from sklearn.metrics import classification_report

report = classification_report(y_test, y_pred)
print("Classification Report:\n", report)
```
